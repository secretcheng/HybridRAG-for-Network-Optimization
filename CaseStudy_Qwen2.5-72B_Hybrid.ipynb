```python
from llama_index.core import Settings, StorageContext, VectorStoreIndex, SimpleKeywordTableIndex, QueryBundle
from llama_index.core.schema import NodeWithScore
from llama_index.core.chat_engine import  CondensePlusContextChatEngine
from llama_index.core.retrievers import (
    BaseRetriever,
    VectorIndexRetriever,
    KeywordTableSimpleRetriever,
    KnowledgeGraphRAGRetriever
)
from llama_index.core.memory import ChatMemoryBuffer
from llama_index.graph_stores.neo4j import Neo4jGraphStore
import qdrant_client
from llama_index.vector_stores.qdrant import QdrantVectorStore
from llama_index.embeddings.ollama import OllamaEmbedding
from llama_index.llms.ollama import Ollama

from typing import List, Any, Optional
from IPython.display import display, Markdown
```

```python
## Load LLM and embeddings

Settings.llm = Ollama(model="qwen2.5:72b", temperature=0.85, context_window=8192, request_timeout=300)
Settings.embed_model = OllamaEmbedding(model_name="bge-m3:latest")
Settings.chunk_size = 1024
```

```python
# 设置聊天历史的 token 限制
memory = ChatMemoryBuffer.from_defaults(token_limit=500000)

SYSTEM_PROMPT = """ 
You are an advanced GraphRAG-powered language model designed to assist in optimizing UAV task scheduling for multi-UAV-assisted mobile edge computing networks. 
Your goal is to reduce energy consumption while addressing carbon emission optimization problems. Leverage your ability to combine graph-based reasoning, contextual awareness, and domain-specific expertise to provide precise, actionable insights.

You must:

-Incorporate principles of retrieval-augmented generation (RAG) and utilize graph-enhanced reasoning to analyze and solve task offloading scenarios.
-Address issues of energy efficiency, task prioritization, and resource allocation within UAV networks.
-Generate structured and explainable outputs, including graphs, tables, and step-by-step solutions, while being mindful of UAV-specific constraints such as limited battery capacity, real-time processing requirements, and environmental factors.
-Ensure scalability and adaptability in your suggestions, enabling effective operation in dynamic environments with multiple UAVs and changing network conditions.

For every query or task:

-Retrieve relevant external knowledge where necessary and integrate it seamlessly into your response.
-Avoid common pitfalls such as flat retrieval or lack of context in dynamic scenarios.
-Prioritize solutions that enhance real-time decision-making, minimize latency, and reduce carbon footprints.
-The mathematical output must be displayed using markdown source code.

"""
```

```python
## Load the graph from generated data
from rag import HybridRAG

Hybrid_retriever = HybridRAG(
    qdrant_host="localhost",
    qdrant_port=6333,
    collection_name="UAV",
    similarity_top_k=3,
    verbose=True,
    )


chat_engine = CondensePlusContextChatEngine.from_defaults(
    retriever=Hybrid_retriever,
    system_prompt=SYSTEM_PROMPT,
    memory=memory,
) 
```

```python
## perform query
response = chat_engine.chat(
    "Hi, you are an expert in carbon emission reduction. Please help me formulate an optimization problem in minimizing carbon emissions. I will show you the scenario later. Based on your local document and do it step-by-step. Output mathematical formulas or symbols must be displayed using markdown source code. If clear, reply me ok and wait for the following instruction.",
)
display(Markdown(f"<b>{response}</b>"))
```

```python
## perform query
response = chat_engine.chat(
    "Good! My scenario is task offloading in multi-UAV-assisted MEC networks. The continuous system timeline T is divided into N time slots. At each time slot t, users can offload computing tasks to UAVs equipped with MEC servers. Can you firstly describe this scenario for me? Based on your local knowledge.",
)
display(Markdown(f"<b>{response}</b>"))
```
